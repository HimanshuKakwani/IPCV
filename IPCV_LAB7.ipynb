# Read Image

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = cv2.imread('img.jpg', 0)
cv2_imshow(img)
```

# Eq Table

```python
n = []
nk = []
pdf = []
cdf = [0]
c = 0
cdf255 = []
op=[]

for i in range(0,256):
  n.append(i)
  nk.append(np.count_nonzero(img == i))

s = sum(nk)

for i in range(0,256):
  pdf.append(nk[i]/s)

for i in range(1,256):
  c = c + pdf[i-1]
  cdf.append(c)

for i in range(0,256):
  cdf255.append(cdf[i]*255)

for i in range(0,256):
  op.append(math.ceil(cdf255[i]))
```

```python
for i in range(160,170):
  print(n[i] , nk[i], pdf[i], cdf[i], cdf255[i], op[i])
```

# Histogram Eq Image

```python
hist_equalized = np.zeros(img.shape)

h,w = img.shape

for i in range(h):
  for j in range(w):
    r = img[i][j]
    hist_equalized[i][j] = op[r]
```

```python
cv2_imshow(hist_equalized)
```

# Input vs Output Histogram

```python
plt.hist(img, bins=7)
plt.show()
```

```python
plt.hist(hist_equalized, bins=7)
plt.show()
```
