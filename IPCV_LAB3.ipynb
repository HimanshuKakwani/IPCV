# Contrast streching

```python
lcimg = cv2.imread('/content/img.jpeg',0)
```

```python
import numpy as np
nstreched = np.zeros(lcimg.shape)
h1,w1 = nstreched.shape
```

```python
a= 50
b = 150
l = 0.5
m = 2
n = 0.3

v = l*a
w = m*(b-a)+l*a

i = 0
j = 0
```

```python
for i in range(h1):
  for j in range(w1):
    r = lcimg[i][j]
    if(r>0 and r<=a):
      nstreched[i][j] = l*(r)
    elif(r>a and r<=b):
      nstreched[i][j] = m*((r)-a)+v
    elif(r>b and r<=255):
      nstreched[i][j] = n*((r)-b)+w
```

```python
cv2_imshow(lcimg)
```

```python
cv2_imshow(nstreched)
```

```python
ir_max = lcimg.max()
ir_min = lcimg.min() 
print(ir_max,ir_min)
is_max = nstreched.max()
is_min = nstreched.min() 
print(is_max,is_min)
```

```python
ir_max = lcimg.max()
ir_min = lcimg.min() 
print(ir_max,ir_min)
is_max = n1streched.max()
is_min = n1streched.min() 
print(is_max,is_min)
```

# Log transformation

```python
img = cv2.imread('/content/download (1).jpeg',0)
```

```python
import math
```

```python
log = np.zeros(img.shape)
h2,w2 = log.shape
c = 50
```

```python
for i in range(h2):
  for j in range(w2):
    r = img[i][j]
    log[i][j] = c*math.log(r+1)
```

```python
cv2_imshow(img)
```

```python
cv2_imshow(log)
```

# Power Transformation

```python
power = np.zeros(img.shape)
h3,w3 = power.shape
g1 = 1.1
g2 = 0.9
```

```python
for i in range(h3):
  for j in range(w3):
    r = img[i][j]
    power[i][j] = r**g1
```

```python
cv2_imshow(power)
```

```python
power2 = np.zeros(img.shape)
h3,w3 = power2.shape
```

```python
for i in range(h3):
  for j in range(w3):
    r = img[i][j]
    power2[i][j] = r**g2
```

```python
cv2_imshow(power2)
```
