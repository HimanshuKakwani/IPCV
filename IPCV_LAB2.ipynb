# Import and Grayscaling

```python
path = r'/content/download (1).jpeg'

import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread(path,cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(image)
```

```python
cv2_imshow(img)
```

# Digital Negative

```python
negative_img = 255 - img
cv2_imshow(negative_img)
```

# Thresholding

```python
h , w = img.shape
print(h,w)
```

```python
import numpy as np
threshold_img = np.zeros(img.shape)
```

```python
t = 125
i = 0
j = 0
for i in range(h):
  for j in range(w):
    if(img[i][j]>=t):
      threshold_img[i][j] = 255
    else:
      threshold_img[i][j] = 0
```

```python
cv2_imshow(threshold_img)
```

```python
cv2_imshow(img)
```

# Gray level slicing

```python
sliced_wbg_img = np.zeros(img.shape)
sliced_wobg_img = np.zeros(img.shape)
```

```python
a1= 100
b1 = 200
i = 0
j = 0
for i in range(h):
  for j in range(w):
    if(img[i][j]>=a1 and img[i][j]<=b1):
      sliced_wbg_img[i][j] = 255
    else:
      sliced_wbg_img[i][j] = img[i][j]
```

```python
a2= 100
b2 = 200
i = 0
j = 0
for i in range(h):
  for j in range(w):
    if(img[i][j]>=a2 and img[i][j]<=b2):
      sliced_wobg_img[i][j] = 255
    else:
      sliced_wobg_img[i][j] = 0
```

```python
cv2_imshow(sliced_wobg_img)
```

```python
cv2_imshow(sliced_wbg_img)
```
